# Base de datos propinas

* objeto a estudiar: cantidad de propinas.
    - `tips`: Cantidad de propinas
* Atributos de nuestra base de datos:
    - `Sex`: Sexo del camarero/a. String
    - `Total_bill`: Total Cuenta
    - `Smoker`: Cantidad fumadores
    - `Day`: Día de atención
    - `Time`: Almuezo o Cena
    - `Size`: Cantidad de personas por mesa

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as statm
import statsmodels.formula.api as frm
plt.style.use('seaborn')
```

```python
# Vamos a importar nuestro cvs
df = pd.read_csv('./tips.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.sex.value_counts() / len(df)
```

```python
df.smoker.value_counts() / len(df)
```

```python
df.day.value_counts() / len(df)
```

```python
df.time.value_counts() / len(df)
```

```python
df.sex.replace(['Female', 'Male'], [1, 0], inplace=True)
```

```python
df.head()
```

```python
df.smoker.replace(['No', 'Yes'], [1, 0], inplace=True)
```

```python
df.head()
```

```python
df.time.replace(['Dinner', 'Lunch'],[1, 0], inplace=True)
```

```python
df.head()
```

```python
df.assign(weekend = df.day.map({'Sat': 1, 'Sun': 1, 'Thur': 0, 'Fri': 0}))
```

```python
df.total_bill.plot(kind = 'hist')
```

```python
df.total_bill.mean()
```

```python
df.total_bill.std()
```

## Ahora vamos por los gráficos

```python
# Cantidad de personas por sexo
cross = pd.crosstab(df['sex'], df['size'])
cross.plot(kind = 'bar')
```

```python
# Cantidad de fumadores por sexo
cross = pd.crosstab(df['sex'], df['smoker'])
cross.plot(kind = 'bar')
```

```python
matrix_corr = df.corr()
matrix_corr
```

```python
sns.heatmap(matrix_corr, annot=True)
```

```python
# Todos los graficos
sns.pairplot(df)
```

```python
# Se carga la variable
model_1 = frm.ols('tip ~ total_bill', data=df)
```

```python
# Con fit corres la migración
model_1 = model_1.fit()
```

```python
model_1.summary()
```
